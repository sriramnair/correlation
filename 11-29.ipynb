```python
import pandas as pd
import scipy as sp
from scipy.stats import pearsonr, spearmanr
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/shnai/OneDrive/Desktop/Data_Specialization/Oil and Gas 1932-2014.csv')
```

```python
data
```

```python
data['mult_nom_2000'] = data['mult_nom_2000'].apply(lambda x: round(x, 4))
```

```python
np.corrcoef(data['mult_nom_2000'], data['mult_2000_2014'])
```

```python
plt.figure(figsize=(12,12))
plt.scatter(x=data['mult_nom_2000'], y=data['mult_2000_2014'], color='m')
plt.xlabel('mult_nom_2000')
plt.ylabel('mult_2000_2014')
plt.title('X vs Y')
plt.show()
```

The correlation shows if there is a direct relationship between two variables. As in a change in one variable will affect
how the other variable behaves. 

```python
pearsonr(df['Attack'], df['Defense'])
```

```python
plt.figure(figsize=(15,15))
plt.scatter(df['Attack'], df['Defense'])
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack vs Defense')
plt.show()
```

```python
spearmanr(df['Speed'], df['Sp. Atk'])
```

```python
plt.figure(figsize=(15,15))
plt.scatter(df['Speed'], df['Sp. Atk'], color='g')
plt.xlabel('Speed')
plt.ylabel('Sp.Atk')
plt.title('Speed vs Sp.Atk')
plt.show()
```
